```python
import requests
import json

base_url = "https://www.kia.ru/ajax/page/accessories/filter?sort=sort&order=desc&page="


headers = {
    "Referer": "https://www.kia.ru/service/accessories/",  # Example Referer header
}

ids_for_urls = []

start_page = 1
end_page = 23

```

```python
for page_number in range(start_page, end_page + 1):
    current_url = base_url + str(page_number)
    response = requests.get(current_url, headers=headers)
    json_data = response.text
    data = json.loads(json_data)



    ids = [tech['article'] for tech in data['content']['accessories']]


    print(ids)

    for id in ids:
      ids_for_urls.append(id)
```

```python
print(ids_for_urls)
```

```python
urls = []
base_url = 'https://www.kia.ru/service/accessories/'

for id in ids_for_urls:
  urls.append(base_url+id)
```

```python
text =[]
```

```python
def has_text_class(tag):
  return tag.has_attr('class') and ('text-s1' in tag['class'] or 'text-s2i' in tag['class'] or 'features' in tag['class'] or 'characteristic' in tag['class'])
```

```python
for url in urls:
  content = requests.get(url)
  if content.status_code == 200:
    #constructor-block

    html = ''.join(line.strip() for line in content.text.split("\n"))
    soup = BeautifulSoup(html, "html.parser")
    name = soup.find('title').text.split(' – ')[0]
    dir = url.split('www.kia.ru')[1].split('/')[1]

    # Extract text from <h1> tag
    h1_text = soup.find('h1').text.strip()
    text.append(h1_text )

    # Extract text from <div> with class "accessory-detail-header__top-article"
    article_text = soup.find('div', class_='accessory-detail-header__top-article').text.strip()
    text.append(article_text)

    img_tag = soup.find('img')


    if img_tag:
      img_url = img_tag.get('data-src')
      text.append('Фото по ссылке ' + img_url)




    text_elements = soup.find_all(has_text_class)

    for text_item in text_elements:
      text.append(text_item.text)


    #print(soup)
```
